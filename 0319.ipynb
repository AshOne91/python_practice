```python
from glob import glob
from tqdm import tqdm # 진행현황
files = glob('./data/sub1/*.csv')
files[0]
```

```python
pd.read_csv(files[0])
# 상가업소번호와 코드가 들어간 명칭은 제외
new_cols = [col for col in df.columns if '코드' not in col]
new_cols = new_cols[1:11] + new_cols[-2:]
new_df = df[new_cols]
new_df.head(1)
```

```python
total_dfs = []
for file in tqdm(files):
  df = pd.read_csv(file)
  # 상가업소번호와 코드가 들어간 명칭은 제외
  new_cols = [col for col in df.columns if '코드' not in col]
  new_cols = new_cols[1:11] + new_cols[-2:]
  new_df = df[new_cols]
  total_dfs.append(new_df)

# 리스트에 담기 데이터프레임들을 열(column) 기준으로 합친다  axis=1
merged_df = pd.concat(total_dfs,axis=0)
```

```python
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm  # 진행형황
files = glob('./data/sub1/*.csv')
files[6]
df = pd.read_csv(files[6])
```

```python
df.columns
```

```python
df.groupby(['시군구명']).agg({'상호명':'count', '경도' : ['mean', 'std'],'위도':['mean', 'std']}).round(2)
```

```python
# 표준 편차 : 각 값들이  평균으로부터 얼마나 떨어져 있는지 나타내는 지표
summary_status =  df.groupby(['시군구명']).agg({
    '상호명' : 'count',  # 상점 수
    '경도' : ['mean','std'],
    '위도' : ['mean','std']
}).round(2)
summary_status.columns = ['상점 수','평균_경도','경도_표준편차','평균_위도','위도_표준편차']
for category in ['음식','소매']:
  summary_status[f'{category} 업종비율 (%)']  = \
    (df[df['상권업종대분류명'] == category].groupby('시군구명').size() / summary_status['상점 수'] * 100).round(2)

summary_status
```
